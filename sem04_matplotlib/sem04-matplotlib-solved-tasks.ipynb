2. Постройте график функциий `f(x) = x^2` и `f(x) = 10*sin(x)`, а также их суммы для x от -5 до 12. Не забудьте обозначить оси и легенду. 

```python
x = np.linspace(-5, 12, 1000)

plt.plot(x, x ** 2,  label='$x^2$')
plt.plot(x, 10*np.sin(x), label='$10*sin(x)$')
plt.plot(x, 10*np.sin(x) + x ** 2, label='$10*sin(x) + x^2$')

plt.xlabel('x')
plt.ylabel('$f(x)$')

plt.legend();
```

3. Постройте верхнюю половину сферы в 3d с центром в точке $(0, 0, 0)$ и радиусом $R = 1$. 

**Подсказка** Сфера однозначно задается функцией вида $x^2 + y^2 + z^2 = R^2$

```python
def f(x, y):
    return np.sqrt(1 - x ** 2 - y ** 2)

x = np.linspace(-1, 1, 150)
y = np.linspace(-1, 1, 150)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```
