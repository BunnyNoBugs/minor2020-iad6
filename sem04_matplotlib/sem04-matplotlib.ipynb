### Visualizing a Three-Dimensional Function

Подробнее и с большим количеством примеров — [тут](https://jakevdp.github.io/PythonDataScienceHandbook/04.04-density-and-contour-plots.html) и [тут](https://jakevdp.github.io/PythonDataScienceHandbook/04.12-three-dimensional-plotting.html)

```python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)

x = np.linspace(0, 5, 10)
y = np.linspace(0, 5, 10)

X, Y = np.meshgrid(x, y)
```

```python
X.shape
```

```python
Y.shape
```

```python
plt.scatter(X, Y)
```

```python
x = np.linspace(0, 5, 40)
y = np.linspace(0, 5, 50)

X, Y = np.meshgrid(x, y)

Z = f(X, Y)
```

```python
plt.contour(X, Y, Z, colors='black');
```

```python
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar();
```

```python
from mpl_toolkits import mplot3d
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

```python
ax = plt.axes(projection='3d')

# Data for a three-dimensional line
zline = np.linspace(0, 15, 1000)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline, yline, zline, 'gray')

# Data for three-dimensional scattered points
zdata = 15 * np.random.random(100)
xdata = np.sin(zdata) + 0.1 * np.random.randn(100)
ydata = np.cos(zdata) + 0.1 * np.random.randn(100)
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='Greens');
```

```python
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

```python
ax.view_init(60, 35)
fig
```

### Subplots!

Без лишних [слов](https://jakevdp.github.io/PythonDataScienceHandbook/04.08-multiple-subplots.html) 

```python
fig, ax = plt.subplots(2, 3, sharex='col', sharey='row')
```

```python
for i in range(2):
    for j in range(3):
        ax[i, j].text(0.5, 0.5, str((i, j)),
                      fontsize=18, ha='center')
fig
```

```python
fig, ax = plt.subplots(1, 2, sharex='col', sharey='row', figsize=(15, 5))

x = np.linspace(0, 10, 1000)
y_1 = np.sin(x)
y_2 = np.cos(x)

ax[0].plot(x, y_1)
ax[0].set_ylabel('sin(x)')
ax[0].set_xlabel('x')

ax[1].plot(x, y_2)
ax[1].set_ylabel('cos(x)')
ax[1].set_xlabel('x');
```

```python
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 3000).T

# Set up the axes with gridspec
fig = plt.figure(figsize=(6, 6))
grid = plt.GridSpec(4, 4, hspace=0.2, wspace=0.2)
main_ax = fig.add_subplot(grid[:-1, 1:])
y_hist = fig.add_subplot(grid[:-1, 0], xticklabels=[], sharey=main_ax)
x_hist = fig.add_subplot(grid[-1, 1:], yticklabels=[], sharex=main_ax)

# scatter points on the main axes
main_ax.plot(x, y, 'ok', markersize=3, alpha=0.2)

# histogram on the attached axes
x_hist.hist(x, 40, histtype='stepfilled',
            orientation='vertical', color='gray')
x_hist.invert_yaxis()

y_hist.hist(y, 40, histtype='stepfilled',
            orientation='horizontal', color='gray')
y_hist.invert_xaxis()
```
